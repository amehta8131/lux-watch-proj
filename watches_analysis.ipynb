```python
# ============================================================
# WATCH MARKET ANALYSIS — A Data Science Project
# Tools: Python, SQL (pandasql), Pandas, Seaborn, Matplotlib
# ============================================================


# -------- SECTION 1: INITIAL IMPORTS AND DATA LOADING ----------

# --- Importing the Core Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# --- Adding SQL in Python ---
!pip install pandasql -q
import pandasql as psql

# --- Display Settings ---
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.family'] = 'DejaVu Sans'

# --- Load Data ---
# Original URL for the Excel file on GitHub
url_github = "https://github.com/amehta8131/lux-watch-proj/blob/main/Watches_cleaned.xlsx"
# Raw URL for direct download
raw_url = "https://github.com/amehta8131/lux-watch-proj/raw/main/Watches_cleaned.xlsx"

# Download the file to the Colab environment
!wget -O Watches_cleaned.xlsx "{raw_url}"

# Load the data from the local file
df = pd.read_excel("Watches_cleaned.xlsx")

print("✅ Data loaded successfully!")
print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
df.head()
```

```python
# -------- SECTION 2: DATA OVERVIEW ----------

# Here is the count of the rows and columns
# Along with some facts including count of unique brands and models, date and price ranges

print(f"Shape: {df.shape}")
print(f"Unique Brands: {df['brand'].nunique()}")
print(f"Unique Models: {df['model'].nunique()}")
print(f"Price Range: ${df['price'].min():,} - ${df['price'].max():,}")
print(f"Average Price: ${df['price'].mean():,.2f}")
print("")

# Summary Statistics for the price column
print("Summary statistics for the 'price' column:")
display(df['price'].describe())

# Count of null values per column
df.isnull().sum()
```

```python
# -------- SECTION 3: Questions + Answers ----------

# The goal of this analysis is to answer the following business questions:
# 1. Which brands have the highest average price on the secondary market?
# 2. Does watch condition impact resale price?
# 3. Which movement type tends to be most expensive?
# 4. What are the most popular case and bracelet materials?
# 5. Does case material influence price?
# 6. What is the price distribution across all watches?
```

```python
# 1. Which brands have the highest average price on the secondary market?


#SQL Code
q1 = psql.sqldf("""
    SELECT brand,
           COUNT(*) as total_watches,
           ROUND(AVG(price), 2) as avg_price,
           MIN(price) as min_price,
           MAX(price) as max_price
    FROM df
    GROUP BY brand
    ORDER BY avg_price DESC
""")
display(q1)

# Bar graph showing average resale by brand
plt.figure(figsize=(12, 6))
sns.barplot(data=q1, x='brand', y='avg_price', palette='rocket')
plt.title('Average Resale Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=50, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 2. Does watch condition impact resale price?


#SQL Code
q2 = psql.sqldf("""
    SELECT condition,
           COUNT(*) as total_watches,
           ROUND(AVG(price), 2) as avg_price
    FROM df
    GROUP BY condition
    ORDER BY avg_price DESC
""")
display(q2)

# Bar graph showing average price by condition
plt.figure(figsize=(10, 5))
sns.barplot(data=q2, x='condition', y='avg_price', palette='rocket')
plt.title('Average Resale Price by Condition')
plt.xlabel('Condition')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 3. Which movement type tends to be most expensive?


#SQL Code
q3 = psql.sqldf("""
    SELECT movement,
           COUNT(*) as total_watches,
           ROUND(AVG(price), 2) as avg_price
    FROM df
    GROUP BY movement
    ORDER BY avg_price DESC
""")
display(q3)

# Bar graph showing average price per movement type
plt.figure(figsize=(8, 5))
sns.barplot(data=q3, x='movement', y='avg_price', palette='Greens_d')
plt.title('Average Price by Movement Type')
plt.xlabel('Movement Type')
plt.ylabel('Average Price ($)')
plt.tight_layout()
plt.show()
```

```python
# 4. What are the most popular case and bracelet materials?

# Most popular case material
q4a = psql.sqldf("""
    SELECT case_material,
           COUNT(*) as total
    FROM df
    GROUP BY case_material
    ORDER BY total DESC
    LIMIT 10
""")

# Most popular bracelet material
q4b = psql.sqldf("""
    SELECT bracelet_material,
           COUNT(*) as total
    FROM df
    GROUP BY bracelet_material
    ORDER BY total DESC
    LIMIT 10
""")

print('Most Popular Case Materials:')
display(q4a)
print('')
print('Most Popular Bracelet Materials:')
display(q4b)


```

```python
# 5. Does case material influence price?

#SQL Code
q5 = psql.sqldf("""
    SELECT case_material,
           COUNT(*) as total_watches,
           ROUND(AVG(price), 2) as avg_price
    FROM df
    GROUP BY case_material
    ORDER BY avg_price DESC
    LIMIT 10
""")
display(q5)

# Bar graph showing price by case material
plt.figure(figsize=(12, 5))
sns.barplot(data=q5, x='case_material', y='avg_price', palette='rocket')
plt.title('Average Price by Case Material')
plt.xlabel('Case Material')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 6. What is the price distribution across all watches?

#SQL Code
q8 = psql.sqldf("""
    SELECT price,
           CASE
               WHEN price < 1000 THEN 'Under $1,000'
               WHEN price BETWEEN 1000 AND 2999 THEN '$1,000 - $2,999'
               WHEN price BETWEEN 3000 AND 5999 THEN '$3,000 - $5,999'
               WHEN price BETWEEN 6000 AND 9999 THEN '$6,000 - $9,999'
               WHEN price BETWEEN 10000 AND 19999 THEN '$10,000 - $19,999'
               ELSE '$20,000+'
           END as price_range,
           COUNT(*) as total_watches
    FROM df
    GROUP BY price_range
    ORDER BY price ASC
""")
q8


# Bar graph showing price distribution
plt.figure(figsize=(12, 6))
sns.barplot(data=q8, x='price_range', y='total_watches', palette='rocket')
plt.title('Price Distribution Across All Watches')
plt.xlabel('Price Range')
plt.ylabel('Number of Watches')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
